```python
d=[]

def data_summary(x):
    if len(x) == 0:
        print("No data available")
    else:
        print("\n \tData Summary: ")
        print("Total elements: ", len(x))
        print("Minimum value: ", min(x))
        print("Maximum value: ", max(x))
        print("Sum of all values: ",sum(x))
        print("Average value: ", round(sum(x) / len(x), 2))

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

def dataset_stat(d):
    print("\n \tDATASET STATISTIC:")
    print("Minimum value: ", min(d))
    print("Maximum value: ", max(d))
    print("Sum of all values: ",sum(d))
    print("Average value: ", round(sum(d) / len(d), 2))

while True:
    
    print("\n \tWelcome to the Data Analzer and Transformer Program")
    print("\nMain Menu: ")
    print("1.Input Data")
    print("2.Display Data Summary (Built-in Functions)")
    print("3.Calculate Factorial (Recusion)")
    print("4.Filter Data by Threshold (Lambda Function)")
    print("5.Sort Data")
    print("6.Display Dataset Statistice (Return Multiple Values)")
    print("7.Exit Program")

    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        arr = input("Enter data for 1D array (separated by spaces): ")
        d = list(map(int,arr.split()))
        print("Data has been stored successfully!")

    elif choice == 2:
        data_summary(d)

    elif choice == 3:
        a = int(input("Enter The Number to calculate the Factorial: "))
        print("Factorial: ",factorial(a))

    elif choice == 4:
        b=int(input("Enter The Threshold Value to Filter data above this Value: "))
        filter_data=list(filter(lambda h:h >= b,d))
        print(filter_data)

    elif choice == 5:
        print("\n \tChoose sorting option:")
        print("1.Ascending")
        print("2.Descending")
        
        c = int(input("Enter your choice: "))
        if c==1:
            print(sorted(d))
        elif c==2:
            print(sorted(d, reverse=True))
        else:
            print("Invalid choice..")

    elif choice == 6:
        if len(d) == 0:
            print("No data available")
        else:
            dataset_stat(d)

    elif choice == 7:
        print("Thank You For Using Data Analyzer And Transformer Program. Goodbye..!")
        break

    else:
        print("Invalid choice")
```
